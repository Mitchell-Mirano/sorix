## Import Sorix

```python
import numpy as np
from sorix import tensor
from sorix.nn import Relu,Linear,BatchNorm1D
import sorix
```

## Random Dataset

```python
features = 5
classes = 2
examples = 10
```

```python
# create random data
X = tensor(np.random.randn(examples, features))
X
```

## Linear

```python
linear = Linear(features,classes)
linear.W,linear.b
```

```python
linear(X)
```

## Batch Norm

```python
bn = BatchNorm1D(classes)
bn.gamma,bn.beta
```

```python
bn(linear(X))
```

## ReLU

```python
relu = Relu()
relu(X)
```

## Linear + ReLU

```python
relu(linear(X))
```

## Linear + BatchNorm + ReLU

```python
relu(bn(linear(X)))
```

## Without bias

```python
linear = Linear(features,classes,bias=False)
relu(bn(linear(X)))
```

## Using GPU

```python
device = 'gpu' if sorix.cuda.is_available() else 'cpu'
device
```

## Linear + GPU

```python
features = 5
classes = 2
examples = 10

X = tensor(np.random.randn(examples, features), device=device)
linear = Linear(features,classes).to(device)
linear(X)
```

## Batch Norm + GPU

```python
bn = BatchNorm1D(classes).to(device)
bn(linear(X))
```

## ReLU + GPU

```python
relu = Relu()
relu(X)
```

## Linear + ReLU + Batch Norm + GPU

```python
relu(bn(linear(X)))
```
